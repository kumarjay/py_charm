```python
user_rating_mean=np.mean(R, axis=1)
R_demeaned=R- user_rating_mean.reshape(-1,1)
R_demeaned
```

```python
from scipy.sparse.linalg import svds
u, sigma, vT=svds(R_demeaned)
```

```python
sigma=np.diag(sigma)
```

```python
all_user_predicted_rating=np.dot(np.dot(u,sigma),vT)+user_rating_mean.reshape(-1,1)
pred_df=pd.DataFrame(all_user_predicted_rating, columns=df_clean_matrix.columns)
pred_df['reviewerID']=user_df
pred_df.set_index('reviewerID', inplace=True)
pred_df
```
